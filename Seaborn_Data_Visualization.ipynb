```python
import seaborn as sns
print("Seaborn version: ",sns.__version__)
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
# Create a Scatter plot

import matplotlib.pyplot as plt

sns.scatterplot(x = 'total_bill', y = 'tip', data = tips)
plt.title('Scatter plot of Total Bill vs Tip')
plt.show()
```

```python
# Line plot

sns.lineplot(x = 'size' , y = 'total_bill', data = tips)
plt.title('Line Plot of Total Bill by size')
plt.show()
```

```python
sns.lineplot(x = 'total_bill' , y = 'tip', data = tips)
plt.title('Line Plot of Tip by Total Bill')
plt.show()
```

```python
sns.lineplot(x = 'day' , y = 'tip', data = tips)
plt.title('Line Plot of Tip by Day')
plt.show()
```

```python
# Categorical Variables in this dataset

# Sex, Smoker, Day, Time, Size(if range is small) 
```

```python
# Categorical Plots
# Bar plot

sns.barplot(x = 'day' , y = 'total_bill' , data = tips)
plt.title('Bar plot of Total Bill by day')
plt.show()
```

```python
# Box Plot
# Circles that lie outside the boxes and lines are the outliers
# Y axis = 10%ile, 20%ile, 30%ile, 40%ile, 50%ile
sns.boxplot(x = 'day', y = 'total_bill', data = tips)

```

```python
# Violin Plot

sns.violinplot(x = 'day' , y = 'tip', data =tips)
```

```python
sns.violinplot(x = 'day' , y = 'total_bill', data =tips)
```

```python
# Histogram

sns.histplot(tips['total_bill'], bins = 10, kde = True)
```

```python
sns.histplot(tips['total_bill'], bins = 10, kde = False)
```

```python
# KDE Plot

sns.kdeplot(tips['total_bill'], fill = True)
```

```python
# Pairplot - shows relation of every variable with every other variable
# For numerical variable, not for categorical variable

sns.pairplot(tips)
```

```python
# Correlation
# If total_bill increases, tip also increases by 0.67%
corr = tips[['total_bill', 'tip', 'size']].corr()
corr
```

```python
# Heatmap

sns.heatmap(corr, annot = True, cmap = 'coolwarm')
```
